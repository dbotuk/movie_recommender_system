```python
import re
import pandas as pd
from tqdm import tqdm
from imdb import IMDb
```

```python
def fetch_movie_details(movie_title):
    try:
        cleaned_title = re.sub(r'\s*\([^)]*\)(?=\s*\(\d{4}\))', '', movie_title)
        retrieved_movies = ia.search_movie(cleaned_title)
        if retrieved_movies and len(retrieved_movies) > 0:
            movie = retrieved_movies[0]
            ia.update(movie, info=['main'])
            
            directors = None
            cast = None
            runtimes = None
            companies = None
            
            if movie.get('directors'):
                directors = ', '.join([director['name'] for director in movie.get('directors')])
            if movie.get('cast'):
                cast = ', '.join([cast['name'] for cast in movie.get('cast')[:5]])
            if movie.get('runtimes'):
                runtimes = ', '.join(movie.get('runtimes'))
            if movie.get('production companies'):
                companies = ', '.join([company['name'] for company in movie.get('production companies')])
        
        return pd.Series({
            'Directors': directors,
            'Cast': cast,
            'Runtimes': runtimes,
            'Production_companies': companies
        })
    except Exception as e:
        print(f"Error fetching data for movie '{movie_title}': {e}")
        return pd.Series({
            'Directors': None,
            'Cast': None,
            'Runtimes': None,
            'Production_companies': None
        })
```

```python
ia = IMDb()
movies = pd.read_table("../data/movies.dat", sep="::", names=['MovieID', 'Title', 'Genres'], engine='python', encoding='latin1')
```

```python
tqdm.pandas()
movies[['Directors', 'Cast', 'Runtimes', 'Production_companies']] = movies['Title'].progress_apply(fetch_movie_details)
movies
```

```python
movies.dropna(inplace=True)
```

```python
movies.to_csv('../data/movies_extended.csv', index=False)
```
